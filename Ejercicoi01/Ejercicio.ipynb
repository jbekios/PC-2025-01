# Ejercicio Clase 03

Resolver una función cuadrática de la forma:

$$
f(x) = ax^2 + bx + c
$$

Se pide:
1. Pedir al usuario que ingrese los valores de \(a\), \(b\) y \(c\).
2. Calcular las raíces usando la fórmula general.
3. Graficar la parábola con `matplotlib`.

Solución:

## 1. Solicitar los datos al usuario

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

# Solicitar al usuario los coeficientes
a = float(input("Ingrese el valor de a: "))
b = float(input("Ingrese el valor de b: "))
c = float(input("Ingrese el valor de c: "))
```

## 2. Calcular las raíces cuadradas usando la fórmula general

```python
# Calcular el discriminante
discriminante = b**2 - 4*a*c

# Calcular las raíces
raiz1 = (-b + cmath.sqrt(discriminante)) / (2*a)
raiz2 = (-b - cmath.sqrt(discriminante)) / (2*a)

print(f"Las raíces son: {raiz1} y {raiz2}")
```

## 3. Graficar la parábola

```python
# Crear datos para el gráfico
x = np.linspace(-10, 10, 400)
y = a*x**2 + b*x + c

# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'{a}x² + {b}x + {c}')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter([raiz1.real, raiz2.real], [0, 0], color='red', zorder=5, label='Raíces')
plt.title("Gráfico de la función cuadrática")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
